```python
import pandas as pd
```

```python
data = pd.read_csv('df_w_audio_ftotal.csv')
display(data.shape)
data.head()
```

```python
dataspotify = data.select_dtypes(include=['number'])
```

```python
dataspotify
```

```python
# The scale of "proline" is much higher than the scale of many other variables!
# K-Means is a distance based algorithm: we need to scale / normalize:
from sklearn.preprocessing import StandardScaler
transformer = StandardScaler().fit(dataspotify)
dataspotify_sca = transformer.transform(dataspotify)


# Now, all features will have the same weight.
pd.DataFrame(dataspotify_sca,columns=dataspotify.columns).head()
```

```python
from sklearn.cluster import KMeans
```

```python
import numpy as np
K = range(2, 20)
inertia = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=1234)
    kmeans.fit(dataspotify_sca)
    inertia.append(kmeans.inertia_)

import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbow Method showing the optimal k')
```

```python
from sklearn.metrics import silhouette_score
K = range(2, 20)
silhouette = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=1234)
    kmeans.fit(dataspotify_sca)
    silhouette.append(silhouette_score(dataspotify_sca, kmeans.predict(dataspotify_sca)))


plt.figure(figsize=(16,8))
plt.plot(K, silhouette, 'bx-')
plt.xlabel('k')
plt.ylabel('silhouette score')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Slhouette score showing the optimal k')
```

```python
#At first I choose 7 clusters but there were too much song inside one cluster so then i choose 12.
```

```python
kmeans = KMeans(n_clusters=12, random_state=1234)
kmeans.fit(dataspotify_sca)
```

```python
# Predicting / assigning the clusters:
clusters = kmeans.predict(dataspotify_sca)

# Check the size of the clusters
pd.Series(clusters).value_counts().sort_index()
```

```python
clusters
```

```python
data['clusters']=clusters
```

```python
data_clusters = data[['track.name','name','clusters']]
```

```python
data_clusters
```

```python
import pickle
pickle.dump(kmeans, open('kmean.pkl', 'wb'))
kmeans = pickle.load(open('kmean.pkl', 'rb'))

pickle.dump(transformer, open('scaler.pkl', 'wb'))
dataspotify_sca = pickle.load(open('scaler.pkl','rb'))
```
