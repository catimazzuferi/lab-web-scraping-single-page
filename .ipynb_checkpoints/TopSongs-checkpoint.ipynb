```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "http://www.popvortex.com/music/charts/top-100-songs.php"
```

```python
response = requests.get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# soup
```

```python
#soup.select('body > div.container > div:nth-child(4) > div.col-xs-12.col-md-8 > div.chart-wrapper > div')
```

```python
song_info = []

for song in soup.find_all('p', class_='title-artist'):
    song_name = song.find('cite', class_='title').get_text()
    artist_name = song.find('em', class_='artist').get_text()
    song_info.append((song_name, artist_name))

#print("Songs:", song_info)
```

```python
top_songs = pd.DataFrame(song_info, columns=['Song Name', 'Artist Name'])

top_songs
```

```python
import random
def recomendar(top_songs, cancion_ingresada):
    cancion_ingresada = cancion_ingresada.lower()
    top_songs['Song Name'] = top_songs['Song Name'].str.lower()
    if cancion_ingresada in top_songs['Song Name'].values:
        canciones_disponibles = top_songs[top_songs['Song Name'] != cancion_ingresada]
        cancion_recomendada = random.choice(canciones_disponibles['Song Name'].values)
        
        return f"Te recomiendo escuchar '{cancion_recomendada}' de {top_songs.loc[top_songs['Song Name'] == cancion_recomendada, 'Artist Name'].values[0]}."
    else:
        return "Lo siento, no puedo recomendar una canción."


cancion_ingresada = input("Ingresa el nombre de una canción: ")
resultado = recomendar(top_songs, cancion_ingresada)
print(resultado)
```

# API Wrappers

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
```

```python
secrets_file = open("secrets.txt","r")
```

```python
string = secrets_file.read()
```

```python
secrets_dict={}
for line in string.split('\n'):
    if len(line) > 0:
        #print(line.split(':'))
        secrets_dict[line.split(':')[0]]=line.split(':')[1].strip()
```

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

#Initialize SpotiPy with user credentials
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=secrets_dict['clientid'],
                                                           client_secret=secrets_dict['clientsecret']))
```

```python
import pandas as pd
from pandas import json_normalize
```

```python
# we will use our 'first paid music' playlist as an example:
playlist = sp.user_playlist_tracks("spotify", "5xqpyfZyS1DVysoevdVyEn")
```

```python
#playlist
```

```python
playlist.keys() 
```

```python
playlist["items"][0].keys()
```

```python
playlist["total"]
```

```python
len(playlist["items"])
```

```python
playlist['next']
```

```python
from random import randint
from time import sleep

def get_playlist_tracks(playlist_id):
    results = sp.user_playlist_tracks("spotify",playlist_id)
    tracks = results['items']
    while results['next']!=None:
        results = sp.next(results)
        tracks = tracks + results['items']
        sleep(randint(1,3000)/1000) # respectful nap
    return tracks
```

```python
all_tracks = get_playlist_tracks("5xqpyfZyS1DVysoevdVyEn")
len(all_tracks)
```

```python
tracks2 = json_normalize(all_tracks)
```

```python
tracks2.head()
```

```python
tracks2['track.artists']
```

```python
def expand_list_dict2(row):
    df = json_normalize(row['track.artists'])
    df['song_id'] = row['track.id']
    return df

tracks2['artists_dfs'] = tracks2.apply(expand_list_dict2, axis=1)
tracks2['artists_dfs'][3]
```

```python
# now we create a new dataframe with all these artists
artist_df2 = pd.DataFrame(columns=['href', 'id', 'name', 'type', 'uri'])
for mini_df in tracks2['artists_dfs']:
    #display(val)
    artist_df2 = pd.concat([artist_df2, mini_df], axis=0)
    
artist_df2
```

```python
# now we merge (join) the two dataframes, keeping only the fields we need
df_merged2 = pd.merge(left=tracks2,
                    right=artist_df2,
                    how='inner',
                    left_on='track.id',
                    right_on='song_id')
df_merged2.head()
```

```python
df_final2 = df_merged2[['track.name', 'name', 'song_id']]
df_final2
```

```python
chunks = [(i, i+100) for i in range(0, len(df_final2), 100)]
chunks
audio_features_list = []
for chunk in chunks:
    id_list100 = df_final2['song_id'][chunk[0]:chunk[1]]
    audio_features_list = audio_features_list + sp.audio_features(id_list100)
    sleep(randint(1,3000)/1000)
len(audio_features_list)
```

```python
audio_features_df = json_normalize(audio_features_list)
```

```python
audio_features_df.drop_duplicates(inplace=True) # duplicates because some songs have more artists
```

```python
df_w_audio_ft = pd.merge(left=df_final2,
                        right=audio_features_df,
                        how='inner',
                        left_on='song_id',
                        right_on='id')
df_w_audio_ft
```
