```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "http://www.popvortex.com/music/charts/top-100-songs.php"
```

```python
response = requests.get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# soup
```

```python
soup.select('body > div.container > div:nth-child(4) > div.col-xs-12.col-md-8 > div.chart-wrapper > div')
```

```python
song_info = []

for song in soup.find_all('p', class_='title-artist'):
    song_name = song.find('cite', class_='title').get_text()
    artist_name = song.find('em', class_='artist').get_text()
    song_info.append((song_name, artist_name))

print("Songs:", song_info)
```

```python
top_songs = pd.DataFrame(song_info, columns=['Song Name', 'Artist Name'])

top_songs
```

```python
import random
def recomendar(top_songs, cancion_ingresada):
    cancion_ingresada = cancion_ingresada.lower()
    top_songs['Song Name'] = top_songs['Song Name'].str.lower()
    if cancion_ingresada in top_songs['Song Name'].values:
        canciones_disponibles = top_songs[top_songs['Song Name'] != cancion_ingresada]
        cancion_recomendada = random.choice(canciones_disponibles['Song Name'].values)
        
        return f"Te recomiendo escuchar '{cancion_recomendada}' de {top_songs.loc[top_songs['Song Name'] == cancion_recomendada, 'Artist Name'].values[0]}."
    else:
        return "Lo siento, no puedo recomendar una cancin."


cancion_ingresada = input("Ingresa el nombre de una cancin: ")
resultado = recomendar(top_songs, cancion_ingresada)
print(resultado)
```
